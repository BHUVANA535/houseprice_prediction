```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
test_id=test.Id
train.head()
```

```python
def clean(df):

    columns = ['Street', 'Alley', 'LandContour', 'Utilities', 'LandSlope', 'Condition1',
               'Condition2', 'RoofMatl', 'BsmtCond', 'BsmtFinType2', 'BsmtFinSF2', 'Heating',
               'CentralAir', 'Electrical', 'LowQualFinSF', 'KitchenQual', 'GarageQual',
               'PavedDrive', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'PoolQC', 'Fence', 
               'MiscFeature']
    df = df.drop(columns=columns, errors='ignore')
    
    return df
train = clean(train)
test = clean(test)
```

```python
def sp(df):
    target='SalePrice'
    df=df.dropna(subset=[target])
    return df
train=sp(train)
```

```python
train.head()
```

```python
test.head()
```

```python
def na_fill(df):
    for column in df.columns:
        df[column].fillna(df[column].mode()[0], inplace=True)  # Use mode()[0] to get the most frequent value
    return df

train=na_fill(train)
test=na_fill(test)
```

```python
from sklearn import preprocessing
```

```python
def s_n(df):
    encoders = {}
    
    # For training DataFrame
    for column in df.select_dtypes(include=['object']).columns:
        encoder = preprocessing.LabelEncoder()
        # Fill NaN values with a placeholder before fitting
        df[column] = df[column].astype(str).fillna('missing')
        df[column] = encoder.fit_transform(df[column])
        encoders[column] = encoder  # Store the encoder for each column 
    return df
```

```python
train= s_n(train)
```

```python
train.head()
```

```python
test=s_n(test)
```

```python
test.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cor_matrix=train.corr()
#corr() is for finding the correlation of the matrix
Y_X_corr=cor_matrix['SalePrice'].sort_values(ascending=False)
print(Y_X_corr)
#import matplotlib.pyplot as plt
#Y_X_corr.plot(kind='bar')
#plt.title('Correlation with SalePrice')
#plt.xlabel('Features')
#plt.ylabel('Correlation Coefficient')
#plt.xticks(rotation=45)
#plt.axhline(0, color='grey', lw=0.8, ls='--')
#plt.show()
```

```python
threshold=0
features = Y_X_corr[abs(Y_X_corr) > threshold].index.tolist()
features.remove('SalePrice')
print(features)
```

```python
X=train[features]
Y=train['SalePrice']
X_test=test[features]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error


# Initialize the GradientBoostingRegressor
gbr = GradientBoostingRegressor(random_state=42, n_estimators=100, learning_rate=0.1, max_depth=3)

# Train the model
gbr.fit(X_train, Y_train)

# Make predictions
Y_pred = gbr.predict(X_val)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(Y_val, Y_pred)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

print("Root Mean Squared Error with GradientBoostingRegressor:", rmse)
```

```python
# Make predictions on the filtered test set
submission = gbr.predict(X_test)

# Create a DataFrame for submission
submission_df = pd.DataFrame({
    'id': test_id.values,
    'SalePrice': submission
})

# Save to a CSV file
submission_df.to_csv('submission.csv', index=False)

print("Submission file created successfully: submission.csv")
```
